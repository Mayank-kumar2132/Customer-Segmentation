```python
h1=pd.get_dummies(train[["Gender","Ever_Married","Graduated","Profession","Spending_Score","Var_1"]])
h1
```

```python
train=pd.concat([train,h1],axis="columns")
train
```

```python
train=train.dropna()
```

```python
train.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(train['Segmentation'])
a=pd.DataFrame(le.fit_transform(train['Segmentation']))
a.head()
temp=train
temp['Segmentation']=a
```

```python
temp.head()
```

```python
temp=temp.dropna()
```

```python
from sklearn.model_selection import train_test_split
x=temp.drop(columns=['Segmentation'])
y=temp.iloc[:,4]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1)
```

```python
y_train=y_train.dropna()
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
from sklearn.ensemble import RandomForestClassifier
rm=RandomForestClassifier()
rm.fit(x_train,y_train)
y_pred=rm.predict(x_test)
```
